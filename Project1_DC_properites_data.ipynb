```python
#Standard Deviation Methold


dc_df2_20009_no_missing.head()

#dc_df2_20009_no_missing['Price'] = dc_df2_20009_no_missing['Price'].astype(float)
meanPRICE = dc_df2_20009_no_missing['PRICE'].mean()
stdPRICE = dc_df2_20009_no_missing['PRICE'].std()
toprange = meanPRICE + stdPRICE * 1.96
botrange = meanPRICE - stdPRICE * 1.96

newdf = dc_df2_20009_no_missing.copy()  #to not mess up the original df
newdf = newdf.drop(newdf[newdf['PRICE'] > toprange].index)
newdf = newdf.drop(newdf[newdf['PRICE'] < botrange].index)

newdf.head()
```

```python
import statsmodels.formula.api as smf
```

```python
#create linear regression to predict prices
result = smf.ols('PRICE ~ BEDRM + BATHRM + KITCHENS', data=newdf).fit()
result.summary()
```

```python
#remove y-intercept (set it to be 0)
result = smf.ols('PRICE ~ BEDRM + BATHRM + KITCHENS -1', data=newdf).fit()
result.summary()
```
